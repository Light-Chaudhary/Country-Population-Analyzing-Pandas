```python
import pandas as pd

url  = "https://datahub.io/core/population/r/population.csv"
df = pd.read_csv(url)
df.columns = ["Country", "Code", "Year", "Population"]

df2020 = df[df['Year'] == 2020]

top10_2020 = df2020.nlargest(10, "Population")
average_population = df2020['Population'].mean()
greater_average = df2020[df2020['Population'] > average_population] 
filtered_df2020 = df2020[df2020['Country'].str.startswith('I')]
print(top10_2020)
print(average_population)
print(greater_average)
print(filtered_df2020)

top10_2020.to_csv('top10_pop2020.csv', index = False)
```
